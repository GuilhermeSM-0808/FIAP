exercicio 2, calcular os coeficientes angulares e lineares de cada reta de previsao dos valores

```python
Y = laranja[:,0]
X = laranja[:,1]
n = np.size(X)

a_laranja = (n * np.sum(X*Y) - (np.sum(X) * np.sum(Y))) / ((n * np.sum(X**2)) - np.sum(X)**2)
b_laranja = np.mean(Y) - a_laranja * np.mean(X)

y = a_laranja*X + b_laranja
```

```python
plt.plot(Y,X)
plt.plot(y,X)
```

```python
Y = toranja[:,0]
X = toranja[:,1]
n = np.size(X)

a_toranja = (n * np.sum(X*Y) - (np.sum(X) * np.sum(Y))) / ((n * np.sum(X**2)) - np.sum(X)**2)
b_toranja = np.mean(Y) - a_toranja * np.mean(X)

y = a_toranja * X + b_toranja

plt.plot(Y,X)
plt.plot(y,X)
```
