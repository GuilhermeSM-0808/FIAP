### Aula 03 - Multiplicacao arrays

como calcular o coeficiente angular de um grafico (a)

Y = ax + b

a = Coeficiente angular

n = numero de elementos

Y = Moscow

X = Datas

O coeficinete angular pode ser obitido utilizando a equacao:

### a = (n * Soma(X*Y) - Soma(X) * Soma(Y) ) / ( nSoma(X**2) - (Soma(X))**2 )

```python
Y = Moscow
X = datas
n = np.size(Y)
```

```python
a = (n*np.sum(X*Y) - np.sum(X) * np.sum(Y)) / (n * np.sum(X**2) - (np.sum(X))**2)
a
```

Calculando o ceficiente linear (b) em Y = a*x + b

#### b = Media(Y) - a * Media(X)

```python
b = np.mean(Y) - a * np.mean(X)
b
```

```python
y = a*X + b
```

```python
np.linalg.norm(Moscow - y)
```

```python
plt.plot(X,Y)
plt.plot(X,y)
```
