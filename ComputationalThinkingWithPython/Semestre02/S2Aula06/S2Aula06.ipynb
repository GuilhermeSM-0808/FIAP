```python
dados = pd.DataFrame({
    'nome':['Ana','Paulo','Jos','Laura','Joo','Amanda','Hanna'],
    'idade':[18,24,30,22,27,24,30]
})

dados['idade ao quadrado'] = dados['idade'].apply(lambda x: x**2)
dados
```

```python
dados = pd.DataFrame({
    'nome':['Ana','Paulo','Jos','Laura','Joo','Amanda','Hanna'],
    'idade':[18,24,30,22,27,24,30]
})

dados_novos = dados.assign(pais='Brasil')

dados_novos['Pode votar'] = dados_novos['idade'].apply(lambda x: 'pode' if x > 18 else 'no pode')
dados_novos

```

```python
dados = pd.DataFrame({
    'nome':['Ana','Paulo','Jos','Laura','Joo','Amanda','Hanna'],
    'idade':[18,24,30,22,27,24,30]
})

def categorizar(x):
    return 'adulto' if x > 24 else 'jovem adulto'

dados['Fase'] = dados['idade'].apply(categorizar)
dados
```

```python
dados = pd.DataFrame({
    'nome':['Ana','Paulo','Jos','Laura','Joo','Amanda','Hanna'],
    'idade':[18,24,30,22,27,24,30]
})

dados['nome maiusculo'] = dados['nome'].apply(lambda x: x.upper())
dados
```

```python
dados = pd.DataFrame({
    'nome':['Ana','Paulo','Jos','Laura','Joo','Amanda','Hanna'],
    'idade':[18,24,30,22,27,24,30]
})

dados['Inicial'] = dados['nome'].apply(lambda x: x[:1])
dados
```

```python
dados = pd.DataFrame({
    'nome':['Ana','Paulo','Jos','Laura','Joo','Amanda','Hanna'],
    'idade':[18,24,30,22,27,24,30]
})

dados['maior de idade'] = dados['idade'].map(lambda x: x > 18)
dados
```

```python
dados = pd.DataFrame({
    'nome':['Ana','Paulo','Jos','Laura','Joo','Amanda','Hanna'],
    'idade':[18,24,30,22,27,24,30]
})

dados['Pode Votar'] = dados['idade'].apply(lambda x: 'sim' if x > 18 else 'no')
dados
```

```python
dados = pd.DataFrame({
    'nome':['Ana','Paulo','Jos','Laura','Joo','Amanda','Hanna'],
    'idade':[18,24,30,22,27,24,30]
})



dados['idade normalizada'] = dados['idade'].apply(lambda x: (x - dados['idade'].min()) / (dados['idade'].max() - dados['idade'].min()))
dados
```

```python
def fibonacci(n):
    if n<2:
        return 1
    else:
        return fibonacci(n-1)+fibonacci(n-2)
    
for i in range(1, 10 ,1):
    print(fibonacci(i))
```
